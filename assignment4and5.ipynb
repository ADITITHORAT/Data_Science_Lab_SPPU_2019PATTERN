# Data Visualisation 1

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
titanic = sns.load_dataset('titanic')

# Display dataset information
print("Dataset Info:")
print(titanic.info())
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title('Survival Count by Gender')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=titanic, x='class', hue='survived')
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=titanic, x='age', hue='survived', kde=True, bins=30)
plt.title('Survival by Age Distribution')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=titanic, x='class', y='fare', palette='coolwarm')
plt.title('Fare Distribution by Class')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=titanic, x='fare', y='age', hue='survived', palette='viridis')
plt.title('Fare vs Age with Survival Status')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=titanic, x='embark_town', hue='survived')
plt.title('Survival Based on Embarkation Point')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=titanic, x='sibsp', hue='survived', multiple='stack')
plt.title('Survival Rate by Number of Siblings/Spouses Aboard')
plt.show()
```

```python
# Survival based on passenger deck (if deck data exists)
if 'deck' in titanic.columns:
    plt.figure(figsize=(12, 6))
    sns.countplot(data=titanic, x='deck', hue='survived', palette='viridis')
    plt.title('Survival Count by Deck')
    plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns for correlation
numeric_titanic = titanic.select_dtypes(include=['float64', 'int64'])

# Plot heatmap for numeric columns
plt.figure(figsize=(12, 6))
sns.heatmap(numeric_titanic.corr(), annot=True, cmap='Blues')
plt.title('Correlation Heatmap for Titanic Dataset')
plt.show()
```

```python
titanic = titanic.dropna(subset=['age'])
plt.figure(figsize=(12, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
plt.title('Age Distribution by Gender and Survival Status on Titanic')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# ---- 1. Univariate Analysis ---- #
df=titanic
# Survival Count
plt.subplot(2, 3, 1)
sns.countplot(x="survived", data=df, palette="coolwarm")
plt.title("Survival Count")

# Gender Distribution
plt.subplot(2, 3, 2)
sns.countplot(x="sex", data=df, palette="viridis")
plt.title("Gender Distribution")

# Age Distribution
plt.subplot(2, 3, 3)
sns.histplot(df["age"], bins=30, kde=True, color="purple")
plt.title("Age Distribution")

# Passenger Class Distribution
plt.subplot(2, 3, 4)
sns.countplot(x="pclass", data=df, palette="Set2")
plt.title("Passenger Class Distribution")

# Fare Distribution
plt.subplot(2, 3, 5)
sns.histplot(df["fare"], bins=30, kde=True, color="blue")
plt.title("Fare Distribution")

plt.tight_layout()
plt.show()

```

```python
# ---- 2. Bivariate Analysis ---- #

plt.figure(figsize=(12, 6))

# Survival rate by gender
plt.subplot(2, 2, 1)
sns.barplot(x="sex", y="survived", data=df, palette="coolwarm")
plt.title("Survival Rate by Gender")

# Survival rate by class
plt.subplot(2, 2, 2)
sns.barplot(x="pclass", y="survived", data=df, palette="Set2")
plt.title("Survival Rate by Class")

# Age vs Fare with survival
plt.subplot(2, 2, 3)
sns.scatterplot(x="age", y="fare", hue="survived", data=df, palette="coolwarm")
plt.title("Fare vs Age (Survival)")

# Survival by Embarkation Port
plt.subplot(2, 2, 4)
sns.barplot(x="embark_town", y="survived", data=df, palette="magma")
plt.title("Survival Rate by Embarkation Town")

plt.tight_layout()
plt.show()
```
