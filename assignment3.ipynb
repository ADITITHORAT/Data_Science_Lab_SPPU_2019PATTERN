```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
import sklearn.preprocessing
import warnings 
warnings.filterwarnings("ignore")
```

```python
df1=pd.read_csv(r'nba.csv')
df2=pd.read_csv(r'Iris.csv')
```

```python
df1.isnull().sum()
```

```python
df1['Height']=df1['Height'].astype(str).str.replace('-','.')       # replace - with . for futher conversion 
df1['Height']=df1['Height'].astype(float)                     # convert height into float 
```

```python
for col in df1.columns:
    if df1[col].dtypes=="object":
        df1[col].fillna(df1[col].mode()[0],inplace=True)
    else:
        df1[col].fillna(df1[col].median(),inplace=True)                
# fill numerical col with median and categorical by mode 
```

```python
df2
```

```python
bins = [18, 25, 30, 35, 40, float('inf')]
labels = ['18-24', '25-29', '30-34', '35-39', '40+']
df1['Age Group'] = pd.cut(df1['Age'], bins=bins, labels=labels, right=False)
grouped_stats = df1.groupby('Age Group')['Salary'].agg(['mean', 'median', 'min', 'max', 'std'])
print(grouped_stats)
```

```python
college_mode = df1['College'].mode()
print("Mode of College:", college_mode)

salary_mean = df1['Salary'].mean()
print("Mean Salary:", salary_mean)


age_counts = df1['Age'].value_counts()
print("Age Counts:\n", age_counts)

age_group_salary = df1.groupby('Age')['Salary'].mean()
print("Mean Salary by Age:\n", age_group_salary)
```
